```python
from datetime import datetime, timedelta

# Define the range of dates
start_date = datetime(2024, 4, 1, 10, 0)  # April 1, 2024, at 10:00 AM
end_date = datetime(2024, 8, 15, 16, 0)  # April 30, 2024, at 4:00 PM

# Days of the week: Sunday (0) to Thursday (4)
valid_days = {0, 1, 2, 3, 4}

# Step size for each timestamp (1 minute)
step = timedelta(minutes=1)

# Generate timestamps
timestamps = []
current = start_date

while current <= end_date:
    # Filter only the valid days and times
    if current.weekday() in valid_days and 10 <= current.hour < 16:
        timestamps.append(current)
    current += step

# Output the timestamps
# for timestamp in timestamps:
#     print(timestamp)

print(len(timestamps))
```

```python
import pandas as pd
df = pd.DataFrame()
df['datetime'] = timestamps
```

```python
df.to_csv("datetime.csv")
```
